```python
from glob import glob
import pandas as pd
from itertools import product
from IPython.display import display
from collections import defaultdict
import pyperclip
```

```python
files = glob("../logs/**/metrics.csv", recursive=True)
```

```python
dataset_names = ["cars196", "cub200", "stanford_online_products"]
loss_names = ["arcface", "normalized softmax", "contrastive", "triplet", "multi similarity"]

table = defaultdict(dict)
for mode in ["", " bgaugment"]:
    for m, metric in zip(["clean", "corrupted"], ["test_mean_average_precision_at_r", "test_corrupted_mean_average_precision_at_r"]):
        for d, l in product(dataset_names, loss_names):
            # the / selects only the base settings
            filter = f"{d} {l}{mode}/"

            data = [pd.read_csv(f) for f in files if filter in f]
            names = [f.split("/")[4] for f in files if filter in f]

            for name, dat in zip(names, data):
                try:
                    r = dat[metric].tail(5)
                    if m == "clean":
                        table[d + " " + m][l + mode] = f"{r.mean()*100:.2f}"
                    else:
                        table[d + " " + m][l + mode] = f"{r.mean()*100:.2f} ± {r.std()*100:.2f}"
                except:
                    pass
    
table = pd.DataFrame(table)
table = table[sorted(table.columns)]
table.sort_index(inplace=True)
print(metric, mode)
display(table)
```
